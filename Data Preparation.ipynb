##  **Data Preparation and Preprocessing for Tourism Analysis in Ireland**

#### **Importing Necessary Libraries**

We will import the required Python libraries for data manipulation, visualization, and numerical operations.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### **Loading a CSV File into a Pandas DataFrame**

To work with tabular data in Python, we use the **Pandas** library to read CSV (Comma-Separated Values) files. The following code loads a dataset into a Pandas DataFrame.


```python
core_data=pd.read_csv(r"H:\CCT\Git\msc-da-ca1-sem-1-Nandini162003\Dataset\TMQ20.csv")
```

#### **Viewing the First Few Rows of a DataFrame**

After loading a dataset into a Pandas DataFrame, it's essential to inspect the first few rows to understand its structure. The **`head()`** method in Pandas helps achieve this.

```python
core_data.head()
```

#### **Getting Information About the DataFrame Structure**

The **`info()`** method in Pandas provides a quick summary of the DataFrame, giving you an overview of its structure, data types, memory usage, and null value counts. It is especially useful for quickly checking the integrity and completeness of the data before performing any analysis.

```python
core_data.info()
```

#### **Statistical Summary of a DataFrame**

The **`describe()`** method in Pandas provides a statistical summary of the numerical columns in a DataFrame. This is useful for quickly understanding the distribution and key statistics of the dataset.

```python
core_data.describe()
```

#### **Dropping `UNIT` Column from a DataFrame**

In data preprocessing, we need to remove unnecessary columns. The **`drop()`** method in Pandas allows us to delete specific columns from a DataFrame.

```python
core_data = core_data.drop(columns=['UNIT'], axis=1)
```

#### Pivoting a DataFrame in Pandas

Pivoting is a powerful data transformation technique that reshapes a DataFrame from a long format to a wide format. The **`pivot()`** method in Pandas allows us to reorganize data by specifying **index**, **columns**, and **values**.

```python
core_data_pivoted = core_data.pivot(index=['Area of Residence', 'Reason for Journey', 'Quarter'], 
                                          columns='Statistic Label', 
                                          values='VALUE').reset_index()
```

#### Viewing the First Few Rows of the Pivoted DataFrame

After pivoting a DataFrame using **`pivot()`**, it’s important to inspect the transformed data. The **`head()`** method in Pandas allows us to quickly view the first few rows.

```python
core_data_pivoted.head()
```

#### Renaming Columns in a Pandas DataFrame

When working with datasets, column names may be too long or unclear. The **`rename()`** method in Pandas allows us to rename columns for better readability and usability.

```python
core_data_pivoted.rename(columns={"Average Length of Stay by Overseas Travellers to Ireland": "Visitor Stay Avg",
                               "Expenditure by Overseas Travellers in Ireland (excluding fares)": "Visitor Spending",
                               "Overseas Trips to Ireland by Non-Residents, with at least one overnight in Ireland":"Overnight Tourist Trips"}, inplace=True)
```

#### Removing the Name of the Column Index in a Pandas DataFrame

When a DataFrame is **pivoted**, the new column headers may have a name assigned to them (e.g., `'Statistic Label'`). The **`columns.name = None`** method removes this name, ensuring a cleaner dataset.

```python
core_data_pivoted.columns.name = None
```

#### **Extracting Year and Quarter from a Column in Pandas**

When working with time-based data, it’s common to extract specific components like **Year** and **Quarter** from a combined column. The `.str` accessor in Pandas helps perform string operations on DataFrame columns.

```python
core_data_pivoted['Year'] = core_data_pivoted['Quarter'].str[:4]
core_data_pivoted['Quarter'] = core_data_pivoted['Quarter'].str[-1]
```

#### **Viewing the First Few Rows of the Pivoted DataFrame**

After transforming and cleaning the DataFrame, it's essential to check if the modifications were applied correctly. The **`head()`** method in Pandas helps display the first few rows of a DataFrame.

```python
core_data_pivoted.head()
```

#### **Reindexing Columns in a Pandas DataFrame**

Reindexing allows you to rearrange the columns in a DataFrame in a specific order. The **`reindex()`** method is used to reorder or change the columns to a new list.

```python
reindex = ['Year', 'Quarter', 'Area of Residence', 'Reason for Journey', 'Visitor Stay Avg','Visitor Spending','Overnight Tourist Trips']

core_data_pivoted = core_data_pivoted.reindex(columns= reindex)
```

#### **Converting Column Data Types in Pandas**

In Pandas, it's common to convert the data type of a column to ensure consistency and make the data more suitable for analysis. The **`astype()`** method is used for type conversion.

```python
core_data_pivoted["Year"] = core_data_pivoted["Year"].astype(int)
core_data_pivoted["Quarter"] = core_data_pivoted["Quarter"].astype(int)
```

#### **Viewing the First Few Rows of the Pivoted DataFrame**

After performing data transformations such as reindexing, renaming columns, or type conversion, it's important to **inspect the first few rows** to ensure that the changes were applied correctly. The **`head()`** method in Pandas provides a simple way to view the first few rows of a DataFrame.

```python
core_data_pivoted.head()
```

#### **Generating Descriptive Statistics for the Pivoted DataFrame**

After transforming and cleaning the data, it's important to gain insights into its distribution, central tendency, and spread. The **`describe()`** method in Pandas provides summary statistics for numerical columns in a DataFrame.

```python
core_data_pivoted.describe()
```

#### **Getting Information About the DataFrame Structure**

To understand the basic structure of a DataFrame, the **`info()`** method in Pandas provides a concise summary. This method is useful for understanding the **data types**, **memory usage**, and **null value counts** of each column.

```python
core_data_pivoted.info()
```
